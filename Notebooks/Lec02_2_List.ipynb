### 2.3.2. if-else 맛보기 

```python
# 다음에서 if-else 가 하는 일이 무엇인지 생각해보자. 
# 각 라인마다 하는 일이 무엇인지 확인하여, 설명하는 주석을 달라. 
a = [1,2,3,4,5] #
x = 1 # 
if 1 in a : # 
    print('%d 가 a 에 들어있습니다' %x)
else: 
    print('%d 가 a 에 없습니다. ')
```

```python
# 값을 바꿔가면서 if 가 하는 일을 정확히 이해해라. 
a = [1,2,3,4,5] #
x = 1 # 
if 1 in a : # 
    print('%d 가 a 에 들어있습니다' %x)
else: 
    print('%d 가 a 에 없습니다. ')
```
